**Treating Missing Values**

In statistics missing data, or missing values, occur when no data value is stored for the variable in an observation.

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), cbar=False)
```

```python
sns.boxplot(x=df['running_time'])
```

Since the nan values are greater than 25 % we will fill the running time with the mean as there isn't a large number of extreme outliers whereas we will fill the null values of the total seasons and episodes with the median values

```python
# #fill the null values with mean

#calculating the IQR
q1 = df['running_time'].quantile(0.25)
q3 = df['running_time'].quantile(0.75)
iqr = q3 - q1

print(iqr)
```

```python
d = df[(df.running_time > q3 + iqr*1.5) | (df.running_time < q1 - iqr*1.5)]
sns.boxplot(d.running_time)
```

```python
plt.figure(figsize=(16,8))
sns.distplot(d.running_time,kde=True)
```

```python
#removing all the outliers
d = df[(df['running_time'] > q1 - 1.5 * iqr) & (df['running_time'] < q3 + 1.5 * iqr)]
#taking the mean 
mean=d['running_time'].mean()
print(mean)
```

```python
d=df.copy()
d['running_time'].fillna(value=mean, inplace=True)
```

```python
d["running_time"].isnull().sum()
```

```python
d2=d.copy()
d2['running_time'].fillna(df['running_time'].mean())
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.kdeplot(x="running_time", data=d)
sns.kdeplot(data=d2,x='running_time')
plt.legend(labels=["Filled removing outliers","Filled without removing outliers"]);
```

doesnt make a difference so we will simply fill the null values with mean obtained from the iqr analysis

```python
df['running_time'].fillna(mean, inplace=True)
```

```python
#filling the null values with mean
df['episodes'].fillna(df['episodes'].median(), inplace=True)
df['seasons'].fillna(df['seasons'].median(), inplace=True)
```
